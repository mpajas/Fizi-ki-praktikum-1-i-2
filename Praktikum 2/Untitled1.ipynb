```python
from scipy import *
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit

#from sympy import *
#init_printing()

from IPython.display import HTML
from IPython.display import display
```

```python
F = [1.1,1,1.5,1.3,1.2,0.7,0.5,0.8]
x = [119.2,106.1,147.9,135.1,132.5,70.8, 60.5,90]



for i in range(size(x)):
    x[i]=x[i]/100

for f in x:
    print(f)
    
print(x)

pars,cov=curve_fit(lambda x,a,b: a*x+b ,x,F)

a=pars[0]
b=pars[1]

x_fit=linspace(0,2,2001)
y_fit=a*x_fit+b

print("a =", pars[0],"±", sqrt(cov[0,0]))
print("b =",pars[1],"±", sqrt(cov[1,1]))

plt.figure(figsize=(14,8.5))
plt.scatter(x,F, linewidth=4, color="black")

plt.plot(x_fit,y_fit,color="red",linestyle="--")
#plt.xlim(-4,4)
#plt.ylim(-0.03,0.03)

plt.title(r"Graf ovisnosti $F$-$x$", fontsize=22)
plt.xlabel(r"$x$ / $\mathrm{m}$",fontsize=22)
plt.ylabel(r"$F(x)$ / $\mathrm{N}$",fontsize=22)
plt.grid();
print("k = ",pars[0],"±", sqrt(cov[0,0]))
```
