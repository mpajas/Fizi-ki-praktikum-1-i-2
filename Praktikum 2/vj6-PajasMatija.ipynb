## Zadatak 1
Nađite raspodjelu intenziteta svjetlosti uzduž ogibne slike nastale na pukotini (Slika 2). Nacrtajte ovisnost
$I(x)/I_0$ intenziteta o udaljenosti od centralnog maksimuma. Valna duljina svjetlosti je $λ = 632,8 \ \mathrm{nm}$
Položaje $x_n^{\mathrm{min}}$ i $x_n^{\mathrm{max}}$ ogibnih minimuma i maksimuma ne tražite temeljem najniže ili najviše
mjerene točke, već ih vizualno procijenite kao da ste vlastoručno povukli glatku krivulju kroz mjerene točke. Izračunajte širinu pukotine b iz položaja ogibnih minimuma i maksimuma, obavezno koristeći jednu jedinu instancu metode najmanjih kvadrata koju primjenjujete na zajednički skup podataka i za minimume i za maksimume

#### Postav eksperimenta
Divergentna leća žarišne duljine $f=-5 \ \mathrm{cm}$ postavlja se ispred lasera na razmaku otprilike $7,5 \ \mathrm{cm}$, a prije pukotine kroz koju se šalje laserski snop svjetlosti. Kako bi se mogla mjeriti ogibna slika, na udaljenosti $2-3 \ \mathrm{m}$ nalazi se fotoćelija unutar uske pukotine na pomičnom bijelom zastoru. Potrebno je ostvariti ogibnu sliku na polovici zastora te zastor pomicati u smjeru ogibnog maksimuma od početnog položaja i izmjeriti barem prva 2 maksimuma i minimuma s obje strane ogibne slike. Foto ćelija spojena je na voltmetar te pošto je očitani napon $U$ proporcionalan intenzitetu svjetlosti $I$ dati će jednak oblik mjerenja.

#### Što tražimo i kako?
Potrebno je pronaći širinu pukotine $b$ kroz koju se šalje snop svjetlosti lasera valne duljine $λ = 632,8 \ \mathrm{nm}$. Prolaskom svjetlosti kroz pukotinu doći će do vidljivog ogibanja svjetlosti, a samu širinu pukotine moguće je odrediti tako da se za početak pronađe profil intenziteta. Profil intenziteta nalazi se mjerenjem većeg broja točaka koje je potrebno normalizirati najvećom izmjerenom vrijednosti intenziteta nakon čega je potrebno odokativno procijeniti na kojim položajima se nalaze ogibni minimumi i maksimumi $x_n^{\mathrm{min}}$ i $x_n^{\mathrm{max}}$. Njih je potrebno objediniti u jednu metodu najmanjih kvadrata kojom se može odrediti $b$, a koja će biti opisana niže.

Prilikom mjerenja zapamćen je početni položaj zastora te je od svih mjernih točaka oduzeta njegova vrijednost, a nakon mjerenja kada je maksimum bio utvrđen podaci su translatirani tako da se on nalazi na nultom položaju $x = 0 \ \mathrm{cm}$. Za preciznije rezultate prilikom mjerenja položaja korištena je pomična mjerka. Tamna vrijednost napona izmjerena je $U_0 = 9,8 \ \mathrm{mV}$ i oduzeta je od svakog mjerenja.

## <center> Ovisnost $U/U_\mathrm{0}$ o položaju $x$



Nakon procjene ogibnih minimuma i maksimuma bilo ih je potrebno objediniti izrazima:

$$x_n^{\mathrm{min}}\approx n\frac{a \lambda}{b}$$

$$x_n^{\mathrm{max}} \approx \pm \frac{ \left ( 2 |n|+ 1 \right )}{2} \frac{a\lambda}{b}$$

To je jednostavno učinjeno tako da je za ogibne maksimume izračunat član $\frac{ \left ( 2 |n|+ 1 \right )}{2}$ za pripadni $n$ koji predstavlja isto značenje kao $n$ u prvom izrazu pošto izrazi imaju jednak oblik. Tako pripremljeni podaci sortirani su prema pripadnim $n$-ovima u tablici niže, a izbačen je nulti ogibni maksimum.

Taj izraz prepoznamo kao jednadžbu pravca i shvatimo kao položaj $x$ koji ovisi o rednom broju minimuma/maksimuma $n$.

$$x(n) = \frac{a\lambda}{b}n + const.$$

Ako je jednadžba pravca $Y = AX + B$ prepoznajemo u prethodnom izrazu iduće veličine:

$$Y = x$$

$$X = n$$

$$A = \frac{a\lambda}{b}$$

$$B = const.$$

te izražavamo izraz za širinu pukotine


$$b = \frac{a\lambda}{A}$$

kao i pripadnu nepouzdanost ovisnih veličina

$$M_b = \frac{a\lambda}{A} \sqrt{{\frac{M_a}{a}}^2 + {\frac{M_A}{A}}^2}$$

Sve je spremno za provođenje MNK, nakon čega je potrebno samo izračunati $b$ i nepouzdanost $M_b$. U grafu ispod crvenom točkom označena je odbačena mjerna točka koja pripada nultom ogibnom maksimumu. U grafu je jasno vidljivo kako bi ta mjerna točka značajno naginjala pravac i kvarila rezultat MNK. 

$A = 0,007577777778253833 \pm 0,00015879562166795335$

$B = 0,00029999998598704127 \pm 0,0003368564225652687$

$$A = (0,0076 \pm 0,000) \ \mathrm{m}$$

$$B = (0,0003 \pm 0,0003) \ \mathrm{m}$$

$b = (0,00019816631578947367 \pm 0,000005280955324259399) \ \mathrm{m}$

## Zadatak 2
Odredite valnu duljinu svjetlosti lasera pomoću interferencije s Fresnelovom biprizmom. Procijenite nepouzdanosti.

#### Postav eksperimenta
Ispred lasera postavlja se leća žarišne duljine $f = 2 \ \mathrm{cm}$ i Fresnelova biprizma udaljena za otprilike $20 \ \mathrm{cm}$ te se tako projicira interferencijska slika na zastor sa milimetarskim papirom koji je udaljen oko $1-2 \ \mathrm{m}$. Interferencijska slika nastaje zbog Fresnelove biprizme koja slama svjetlost jednog izvora tako da razdvoji snop i daje dojam kao da postoje 2 nova izvora, koje zovemo virtualni izvori. Snopovi svjetlosti tih virtualnih izvora dolaze u interferenciju na zastoru. 

Za drugi dio vježbe potrebno je uvesti leću žarišne duljine $f = 30 \ \mathrm{cm}$ na položaju otprilike $20 \ \mathrm{cm}$ nakon biprizme kako bi se na zastoru realizirala realna slika virtualnih izvora, koji su sada predmet te leće.

#### Što tražimo i kako?
U prvom dijelu zadatka cilj je pronaći razmak između interferencijskih minimuma i maksimuma. Kada se realizira interferencijska slika na milimetarski papir označe se olovkom što je bolje moguće interferencijski minimumi ili maksimumi, nakon čega se pomičnom mjerkom precizno mogu izmjeriti razmaci ocrtanih točaka.

Drugi dio zadatka je potreban kako bi se pronašao razmak između virtualnih izvora i njihova udaljenost od zastora u svrhu određivanja valne duljine lasera $\lambda$ pomoću izraza za razliku optičkih puteva:

$$\Delta = \frac{xd}{a+b}$$

$$\Delta_n^{\mathrm{max}} = n\lambda$$

Pri čemu su veličine:
- $x$ - udaljenost točke na zastoru od koje se promatra interferencija od centralnog maksimuma
- $d$ - razmak između virtualnih izvora
- $(a+b)$ - udaljenost leće od predmeta $a$ i leće od slike $b$, dakle udaljenost predmeta od slike

Kako bi se odredila udaljenost leće od predmeta koristi se relacija leće uz mjerenu veličinu $b$ i poznatu tvornički određenu veličinu $f$:

$$\frac{1}{f}=\frac{1}{a} + \frac{1}{b}$$

i relacija za veličinu predmeta čiju sliku leća stvara pri čemu je $y$ izmjerena veličina slike, odnosno razmak između točaka na zastoru:

$$\frac{a}{b} = \frac{d}{y}$$

U tablici niže mogu se pronaći mjerenja prvih 5 interferencijskih maksimuma od centralnog maksimuma, koji su zatim podijeljeni pripadnim rednim brojem kako bi se što bolje procijenio razmak prvog maksimuma računanjem srednje vrijednosti i nepouzdanosti iz podataka. To se može raditi zato što su interferencijski maksimumi/minimumi ekvidistantni.

### Razmak između leće i virtualnih izvora
Računamo udaljenost virtualnih izvora od leće iz prethodno spomenutog izraza za leću $\frac{1}{f}=\frac{1}{a} + \frac{1}{b}$

$$a = \frac{fb}{b-f}$$

te pripadnu nepouzdanost

$$M_a = \frac{fb}{b-f} \sqrt{\frac{f^2}{b^2(b-f)^2}M_b^2}$$

$a = (0,41842105263157897 \pm 0,0015581717451523544) \ \mathrm{m}$

### Rezultat

$$a = (0,418 \pm 0,002) \ \mathrm{m}$$

### Udaljenost između virtualnih izvora
Sada možemo izračunati veličinu predmeta leće koja predstavlja razmak između virtualnih izvora pomoću izraza $\frac{a}{b} = \frac{d}{y}$

$$d = \frac{ay}{b}$$

uz pripadnu nepouzdanost

$$M_d = \frac{ay}{b} \sqrt{\frac{M_a^2}{a^2} + \frac{M_b^2}{b^2} + \frac{M_y^2}{y^2}}$$

$d = (0,0009473684210526315 \pm 0,00004062629535896141) \ \mathrm{m}$

### Rezultat

$$d = (0,95 \pm 0,04) \times 10^{-3} \ \mathrm{m}$$

### Razlika optičkih puteva
Konačno možemo upotrijebiti sve prethodne podatke kako bismo odredili razliku optičkih puteva

$$\Delta = \frac{xd}{a+b}$$

i pripadnu nepouzdanost

$$\Delta = \frac{xd}{a+b} \sqrt{\frac{M_x^2}{x^2} + \frac{M_d^2}{d^2} + \frac{M_a^2}{(a+b)^2} + \frac{M_b^2}{(a+b)^2}}$$

$\Delta = (548,5226059095762e \pm 27,049116126896107) \times 10^{-9} \ \mathrm{m}$

### Rezultat

$$\Delta = (550 \pm 30) \times 10^{-9} \ \mathrm{m}$$

### Valna duljina

Valna duljina je povezana razlikom optičkih puteva te pošto je određen razmak prvog interferencijskog maksimuma možemo jednostavno uzeti $n=1$ i izraz $\Delta_n^{\mathrm{max}} = n\lambda$ svesti na $\lambda = \Delta$.

Nepouzdanost ovisnih veličina lako se dobije:

$$M_\lambda = \sqrt{\left(\frac{\partial \lambda}{\partial \Delta}\right)^2 M_\Delta^2} = \sqrt{M_\Delta^2} = \left| M_\Delta \right|$$

## Konačni rezultat
$$\lambda = (550 \pm 30) \times 10^{-9} \ \mathrm{m}$$

#### Zadatak 1
U prvom zadatku koristili smo činjenicu da je izmjereni napon proporcionalan intenzitetu svjetlosti što se pokazalo točnim s obzirom na to da smo dobili jednak oblik raspodjele. Koliko je širina pukotine točno određena upitno je bez obzira na nepouzdanost od oko $2 \%$ zbog velikog utjecaja okoline na očitani napon. No s obzirom na to da je postignut dobar oblik ovisnosti intenziteta o pomaku možemo naslutiti da je taj šum ako ništa, bio konzistentan i nije uveo neke drastične greške koje bi nas navele na miskoncepcije.
