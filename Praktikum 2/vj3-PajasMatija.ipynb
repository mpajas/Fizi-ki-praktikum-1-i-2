### Zadatak 2:
Metodom najmanjih kvadrata odredite koeficijent ekspanzije destilirane vode pri temperaturama $20 \ ^{\mathrm{o}}\mathrm{C}$
i $40 \ ^{\mathrm{o}}\mathrm{C}$. Da ne bismo morali uvoditi dodatne pretpostavke o vrijednosti $\rho$ koja stoji pred derivacijom iz
jednadžbe (1), primijetite da se isti izraz može zapisati kao:

$$\alpha=-\frac{\partial \ln \rho}{\partial T}$$

Na kako transformirane toćke ćete stoga primijeniti metodu najmanjih kvadrata (kao i uvijek, u općenitom
obliku y = ax + b) da biste odredili $\alpha$?

Počevši od jednadžbe (1) sa početka pripreme može se malo urediti izraz:

$$
\begin{align}
\alpha &=-\frac{1}{\rho}\frac{\partial \rho}{\partial T} \\ 
-\alpha \mathrm{d} T &= \frac{\mathrm{d} \rho}{\rho}
\end{align}
$$

Te je sada jasno da integriranjem izraza možemo dobiti implicitnu ovisnost $\rho$ o T:

$$\ln \left ( \rho \right ) = -\alpha T + const.$$

Ovu izraz prepoznajemo kao jednadžbu pravca $y=ax + b$ i odredimo pripadne veličine:

$$
\begin{align}
y &= \ln \left ( \rho \right ) \\
x &= T \\
a &= -\alpha \\
b &= const
\end{align}
$$

Nepouzdanost od $\alpha$ jednaka je nepouzdanosti parametra a:
$$
M_{\alpha}=\sqrt{\left ( \frac{\partial \alpha}{\partial a} M_{a}\right )^2}=
\sqrt{\left ( - M_{a}\right )^2}=M_a
$$

Te sada možemo provesti metodu najmanjih kvadrata nad podacima oko $20 \ ^{\mathrm{o}}\mathrm{C}$ i $40 \ ^{\mathrm{o}}\mathrm{C}$.

## Zaključak:

U prvome zadatku najveća gustoća vode izmjerena je na $4 \ ^{\mathrm{o}}\mathrm{C}$ što se definitivno očekuje, ali mjerenja znatno odstupaju od stvarnih vrijednosti gustoće vode. Za usporedbu, kod $4 \ ^{\mathrm{o}}\mathrm{C}$ očekuje se oko $0,999972 \ \mathrm{\frac{g}{{cm}^3}}$ dok je izmjereno $1,0019 \ \mathrm{\frac{g}{{cm}^3}}$,

dok se za $20 \ ^{\mathrm{o}}\mathrm{C}$ očekuje $0,9982071 \ \mathrm{\frac{g}{{cm}^3}}$, a izmjereno je $0,9937 \ \mathrm{\frac{g}{{cm}^3}}$.

Razlog tomu je što vaga nije dovoljno dobro baždarena na početku te se u mjerenje potkrala pogreška uz greške uzrokovane diranjem utega toplomjerom te gibanja vode i preostalih komadića leda.

Uspoređujući izmjerene vrijednosti sa tabličnim podacima jasno se vidi da stvarna vrijednost nije u intervalu nepouzdanosti.

Izmjereno:
$$\alpha_{20^\mathrm{0}\mathrm{C}} = (0,00016 \pm 0,00002) \ \frac{1}{^{\mathrm{o}}\mathrm{C}}$$

$$\alpha_{40^\mathrm{0}\mathrm{C}} = (0,0006 \pm 0,00008) \ \frac{1}{^{\mathrm{o}}\mathrm{C}}$$
