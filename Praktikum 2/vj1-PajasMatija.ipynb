Napetost površine računamo izrazom 

$$\gamma = \frac{F_\mathrm{n}}{l} = \frac{F_\mathrm{n}}{4r\pi} $$

no prije toga potrebno je izračunati nepouzdanost radijusa i radijus. Promjer je izmjeren uz procijenjenu nepouzdanost $d = 19,6 \pm 0,1 \ \mathrm{mm}$.

Radijus je dan izrazom

$$r = \frac{d}{2} = 9,8 \ \mathrm{mm}$$

te njegova nepouzdanost

$$ M_r=\sqrt{  \left ( \frac{\partial r}{\partial d} M_d  \right )^2 }= \left | \frac{M_d}{2} \right | = \frac{M_d}{2} = 0,05 \ \mathrm{mm} $$

Ovim podacima može se sada izračunati nepouzdanost izraza za napetost površine koja glasi

$$
\begin{align}
M_\gamma &=\sqrt{
\left ( \frac{\partial \gamma}{\partial F_\mathrm{d}} M_{F_\mathrm{d}}\right )^2+
\left ( \frac{\partial \gamma}{\partial r} M_{r}\right )^2}\\ M_\gamma &=
\sqrt{
\left ( \frac{1}{4r\pi} M_{F_\mathrm{d}}\right )^2+ \left (- \frac{F_\mathrm{d}}{4r^2 \pi} M_r\right )^2

\sqrt{
\left ( \frac{F_\mathrm{d}}{4r\pi} \cdot \frac{M_{F_\mathrm{d}}}{F_\mathrm{d}}\right )^2+ \left ( \frac{F_\mathrm{d}}{4r \pi} \cdot \frac{M_r}{r}\right )^2
}
\\M_\gamma &=\gamma \cdot \sqrt{
\left ( \frac{M_{F_\mathrm{d}}}{F_\mathrm{d}}\right )^2+ \left (   \frac{M_r}{r}\right )^2
}= 0,0003033689997247447 \ \mathrm{\frac{N}{m}}
\end{align}
$$

$$\gamma = (0,035363253937000204 \pm 0,0003033689997247447) \ \mathrm{\frac{N}{m}}$$

Kako bi se odredili $T_\mathrm{K}$ i $k_\gamma$ koristi se idući izraz koji prepoznajemo kao jednadžbu pravca te sa podacima iz tablice iznad provodi se metoda najmanjih kvadrata.

$$\gamma(T)=-\gamma_0 T+\gamma_0 T_K$$

Ako je jednadžba pravca $y=ax+b$, tada su

$$
\begin{align}
y&=\gamma \\
x&=T \\
a&=-\gamma_0 \\
b&=\gamma_0 T_K\\
\end{align}
$$

Prije nego krenemo sa računanjem podataka, prepoznajmo potrebne izraze i njihove nepouzdanosti:

### Izraz i nepouzdanost za $\gamma_0$

$$\gamma_0=-a$$

$$
M_{\gamma_0}=\sqrt{\left ( \frac{\partial \gamma_0}{\partial a} M_{a}\right )^2}=
\sqrt{\left ( - M_{a}\right )^2}=\left | -M_{a} \right | = M_{a}
$$


### Izraz i nepouzdanost za $T_\mathrm{K}$

$$ T_K=\frac{b}{\gamma_0} $$

$$
\begin{align}
M_{T_K}&=\sqrt{\left ( \frac{\partial T_K}{\partial b} M_{b}\right )^2+
\left ( \frac{\partial T_K}{\partial \gamma_0} M_{\gamma_0}\right )^2
}\\M_{T_K}&=
\sqrt{
\left ( \frac{1}{\gamma_0} M_{b}\right )^2+
\left (- \frac{b}{ \gamma_0^2} M_{\gamma_0}\right )^2
}\\M_{T_K}&=
\sqrt{
\left ( \frac{b}{\gamma_0} \cdot \frac{M_{b}}{b}\right )^2+
\left ( \frac{b}{ \gamma_0} \cdot \frac{M_{\gamma_0}}{\gamma_0}\right )^2
}\\M_{T_K}&=
T_K \cdot \sqrt{
\left ( \frac{M_{b}}{b}\right )^2+
\left ( \frac{M_{\gamma_0}}{\gamma_0}\right )^2
}
\end{align}
$$

### Izraz i nepouzdanost za $k_\gamma$

$$k_\gamma=\gamma_0V_m^{2/3}$$

gdje je $V_m = 18 \times 10^{-6} \ \mathrm{m^3}$ te se tretira kao konstanta.

$$
\begin{align}
M_{k_\gamma}&=\sqrt{\left ( \frac{\partial k_\gamma}{\partial \gamma_0} M_{\gamma_0}\right )^2}
\\M_{k_\gamma}&=\sqrt{\left (V_m^{2/3} \cdot M_{\gamma_0}\right )^2}
\\M_{k_\gamma}&=\sqrt{\left ( \gamma_0 V_m^{2/3}  \cdot \frac{M_{\gamma_0}}{\gamma_0}\right )^2}=k_{\gamma} \cdot \frac{M_{\gamma_0}}{\gamma_0}
\end{align}
$$

### Rezultat metode najmanjih kvadrata

$a = (-0,00010263590646685444 \pm 0,000007777263325090237) \ \mathrm{\frac{N}{Km}}$

$a = (-103 \pm 8) \times 10^{-6} \ \mathrm{\frac{N}{Km}}$

$b = (0,06225828871081196 \pm 0,0026166762000309936) \ \mathrm{\frac{N}{m}}$

$b = (0,062 ± 0,003) \ \mathrm{\frac{N}{m}}$

Te se sada prepoznaju ili izračunaju tražene veličine. Prvo se prepoznaje $\gamma_0$

$$\gamma_0 = (0,103 \pm 0,008) \times 10^{-3} \ \mathrm{\frac{N}{Km}}$$

nakon čega se može izračunati $T_\mathrm{K}$ i njegova nepouzdanost

$$ T_K=\frac{b}{\gamma_0} = 601,9417475728155 \ \mathrm{K}$$

$$M_{T_K}= 55,08317904787849 \ \mathrm{K}$$

te $k_\gamma$ i njegova nepouzdanost

$$k_\gamma=\gamma_0V_m^{2/3} = 7,074334018979595 \times 10^{-8} \ \mathrm{\frac{J}{K}} $$

$$M_{k_\gamma} = 5,494628364255997 \times 10^{-9} \ \mathrm{\frac{J}{K}}$$

$T_K = (600 \pm 60) \ \mathrm{K} $

### Zaključak

#### Zadatak 1
Tablični podatak za iznos površinske napetosti vode je $\gamma = 72.80 \times 10^{-3} \ \mathrm{\frac{N}{m}}$ dok je izmjereni $\gamma = (35,4 \pm 0,3) \times 10^{-3} \ \mathrm{\frac{N}{m}}$ što je gotovo pola. #### Zadatak 2
Tablični podatak za temperaturni koeficijent je $\gamma_0 = 0.1514 \times 10^{-3} \ \mathrm{\frac{N}{Km}}$ dok je izmjereni $\gamma_0 = (0,103 \pm 0,008) \times 10^{-3} \ \mathrm{\frac{N}{Km}}$ koji je puno bliži tabličnoj vrijednosti nego prethodni zadatak. Već je samo mjerenje sa vrućom vodom bilo dovoljno turbulentno, tako da su se spomenute smetnje iz prvog zadatka "prigušile". Osim toga treba primijetiti kako je dobiveni $k_\gamma = (0,71 \pm 0,05) \times 10^{-7} \ \mathrm{\frac{J}{K}} < 2,1 \times 10^{-7} \ \mathrm{\frac{J}{K}}$ što sugerira da molekule vode asociraju s porastom temperature.
